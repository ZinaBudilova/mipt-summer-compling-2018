# Инструменты для обработки и визуализации данных

В данном курсе мы будем использовать самые популярные библиотеки для работы с датасетами: pandas и matplotlib, а также математическую библиотеку -- numpy

Если вы работаете в среде jupyter, то эти библиотеки скорее всего установлены и следующий код выполнится без ошибок

```python
import numpy  as np  # быстрые математические операции
import pandas as pd # для считывания данных
import matplotlib.pyplot as plt # для рисования графиков
```

```python
plt.plot([1, -1, 3, 4, 10], marker = 'o')
plt.show()
```

```python
plt.plot([1, 2, 3, 2, 1])
plt.plot([2], marker= '*')
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
xs = range(10) # 0, 1, ..., 99
ys = [x**2 for x in xs] # 0, 1, 4, 9, 16, ..., 99*99
plt.plot(xs, ys, label='x^2')
plt.plot(xs, xs, label='x')
plt.legend() # если не вызвать, то легенда не появится
plt.show()
```

```python
xs = np.arange(0, 20, 0.1) # массив чисел от 0 до 20 с шагом 0.1
ys = np.sin(xs) # синус от каждого элемента массива
plt.plot(xs, ys)
plt.show()
```

```python
xs = np.arange(0.1, 5, 0.1)
ys = np.log(xs)
plt.plot(xs, ys, marker = '*')
plt.show()
```

```python
xs = np.random.normal(size = 30000)
ys = np.random.normal(size = 30000)
```

```python
plt.scatter(xs, ys, marker = ',', s = 1) 
plt.show()
```

```python
plt.hist(xs, bins=32)
plt.show()
```
