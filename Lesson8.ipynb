# Генераторы

```python
for x in map(lambda x: x**2, range(909090000)):
    if x > 100:
        break
    print(x)
```

генератор -- функция которая может продолжать работать дальше после того как вернула значение

```python
def counter():
    a = 0
    while True:
        yield a
        print('getting new value')
        a += 1
```

```python
g = counter()
print(next(g))
print(next(g))
print(next(g))
```

```python
for i in counter():
    print(i)
    if i > 5:
        break
```

чтобы генератор в цикле for работал не бесконечно он должен в какой-то момент вызвать return (в более ранних версиях python бросить исключение `StopIteration`)

```python
def counter(limit):
    a = 0
    while a < limit:
        yield a
        a += 1
    return
```

```python
g = counter(3)
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

```python
g = map(lambda x: x**2, range(100))
next(g)
next(g)
next(g)
```

```python
for i in counter(10):
    print(i)
```

Упражнение: напишите при помощи генератора аналог функции `range`, котоая поддерживает бесконечные последовательности

например

`myrange(10) -> 1..9`

`myrange() -> 1..`

```python
def myrange (x = 0):
    a = 0
    if x == 0:
        while True:
            yield a
            a += 1
    else:
        while a < x:
            yield a
            a += 1
```

```python
for i in myrange():
    print(i)
```

```python
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
```

```python
# Печатает все числа последовательности Фибоначчи, меньше 1000
for a in fibonacci_generator():
    if a > 1000:
        break
    print(a)
```

Упражнение: напишите генератор выдающий бесконечную последовательность факотриалов

```python
def infin_f():
    x = 1
    res = 1
    while True:
        res *= x
        yield res
        x += 1
```

```python
for a in infin_f():
    if a > 1000:
        break
    print(a)
```

```python
def myzip(a, b):
    while True:
        x = (next(a), next(b))
        yield x
```

```python
a = [1, 2, 3, 4]
b = ['x', 'y', 'z']
y = myzip(iter(a),iter(b))
```

```python
next(y)
```

```python
def myenumerate(a):
    i = 0
    while True:
        x = (i, next(a))
        yield x
        i += 1
```

```python
a = ['x', 'y', 'z']
z = myenumerate(iter(a))
```

```python
next(z)
```

```python
def trigram(a):
    s = ''
    x = next(a) 
    y = next(a)
    xy = x+y
    while True:
        z = next(a)
        s += xy + z
        yield s
        s = ''
        xy = y + z
```

```python
a = 'Буря мглою небо кроет, вихри снежные крутя'
z = trigram(iter(a))
```

```python
print(next(z))
print(next(z))
```

# модуль itertools

```python
from itertools import product, permutations
```

```python
for p in product('abcd', (1, 2, 3, 4)):
    print(p)
```

```python
for n in product((0, 1), repeat=5):
    print(n)
```

```python
for p in permutations([1,2,3]):
    print(p) 
```
